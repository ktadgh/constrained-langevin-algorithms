```python
def graph_points(q,t, alpha=0.1):
  xs = q[:, 0]
  ys = q[:, 1]
  zs = q[:, 2]

  # Create a new figure
  fig = plt.figure()
  plt.title(f't = {t}')
  ax = fig.add_subplot(111, projection='3d')

  # Plot the data
  ax.scatter(xs.cpu(), ys.cpu(), zs.cpu(), alpha = alpha, s = 1)

  # Show the plot
  plt.show()
```

```python
qs = torch.load('../data/fire_qps.pth')
qs = qs.reshape(-1,3)
```

```python
graph_points(qs,0,alpha=0.5)
```

```python
qs = torch.load('../proper-fire-training/my_qs_1500.pt').detach()
qs = qs.reshape(-1,3)
graph_points(qs,0,0.5)
```
